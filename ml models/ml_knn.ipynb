```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt


# sample data
data = {
    'Customer': [
        'Riya', 'Aman', 'Faizan', 'Neha', 'Imran',
        'Sofia', 'Raj', 'Priya', 'Karan', 'Anjali',
        'Vikram', 'Sara', 'Arjun', 'Meera', 'Kabir',
        'Nisha', 'Rahul', 'Pooja', 'Zaid', 'Isha'
    ],
    'Age': [
        20, 30, 40, 22, 38,
        27, 35, 29, 31, 24,
        45, 26, 33, 28, 37,
        23, 41, 32, 36, 25
    ],
    'Spending': [
        100, 200, 300, 110, 290,
        180, 250, 210, 230, 150,
        320, 170, 260, 190, 280,
        140, 310, 240, 270, 160
    ]
}


df = pd.DataFrame(data)

X = df[["Age",'Spending']]  #Pick only the numeric features for clustering
# random_state = [10,15,25,42] # here we can run for different random states but will take time or we can use like loop through it and then save the files in a text file 

model = KMeans(n_clusters=2, random_state=42 , n_init=10)

df['Group'] = model.fit_predict(X)

plt.figure(figsize=(6,5))

for group in df['Group'].unique():
    group_data =  df[df['Group']==group]
    plt.scatter(group_data['Age'],group_data['Spending'], label=f'Group {group}')
plt.xlabel('Age')    
plt.ylabel('Spending Score')
plt.title('Customer Segments(K-Means)')
plt.legend()
plt.grid(True)
plt.show()
print(df)    
```

```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

data = {
    'Customer': [
        'Riya', 'Aman', 'Faizan', 'Neha', 'Imran',
        'Sofia', 'Raj', 'Priya', 'Karan', 'Anjali',
        'Vikram', 'Sara', 'Arjun', 'Meera', 'Kabir',
        'Nisha', 'Rahul', 'Pooja', 'Zaid', 'Isha'
    ],
    'Age': [
        20, 30, 40, 22, 38,
        27, 35, 29, 31, 24,
        45, 26, 33, 28, 37,
        23, 41, 32, 36, 25
    ],
    'Spending': [
        100, 200, 300, 110, 290,
        180, 250, 210, 230, 150,
        320, 170, 260, 190, 280,
        140, 310, 240, 270, 160
    ],
    'Income': [
        3000, 5000, 7000, 3200, 6800,
        4500, 6200, 4800, 5300, 3600,
        7500, 4400, 6100, 4700, 6900,
        3400, 7200, 5400, 6700, 3800
    ]
}

df = pd.DataFrame(data)

# Kmeans clustering with 3 features 
X = df[['Age','Spending','Income']]

model = KMeans(n_clusters=3,random_state=42,n_init=10)
df['Group']= model.fit_predict(X)

# Redue to 2D using PCA for visualization
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

df['PCA1'] = X_pca[:,0]
df['PCA2'] = X_pca[:,1]

# plotting the clusters in 2D
plt.figure(figsize=(12,6))
colors = ['blue','yellow','red']

for i , group in enumerate(df['Group'].unique()):
    group_data = df[df['Group']== group]
    plt.scatter(group_data['PCA1'],group_data['PCA2'], label = f"Group {group}",color = colors[i], s=50)
plt.xlabel('PCA component 1')    
plt.ylabel('PCA component 2')    
plt.title('Customer Segments (3 Features -> PCA + K-Means)')
plt.grid(True)
plt.show()

print(df)

```
